```python
import pandas as pd # type: ignore
dataset = [
    ['less than 30', 'high', 'no', 'fair', 'no'],
    ['less than 30', 'high', 'no', 'excellent', 'no'],
    ['31-40', 'high', 'no', 'fair', 'yes'],
    ['>40', 'medium', 'no', 'fair', 'yes'],
    ['>40', 'low', 'yes', 'fair', 'yes'],
    ['>40', 'low', 'yes', 'excellent', 'no'],
    ['31-40', 'low', 'yes', 'excellent', 'yes'],
    ['less than 30', 'medium', 'no', 'fair', 'no'],
    ['less than 30', 'low', 'yes', 'fair', 'yes'],
    ['>40', 'medium', 'yes', 'fair', 'yes'],
    ['less than 30', 'medium', 'yes', 'excellent', 'yes'],
    ['31-40', 'medium', 'no', 'excellent', 'yes'],
    ['31-40', 'high', 'yes', 'fair', 'yes'],
    ['>40', 'medium', 'no', 'excellent', 'no']
]
columns = ['Age', 'Income', 'Student', 'CR', 'Buys_Computer']
#list to dataframe converstion
df = pd.DataFrame(dataset, columns=columns)
# Print the DataFrame
print(df)
#Step 2 : prior probablity
# yes = no. of yes/14
# no = no. of no /14
prior = df.groupby('Buys_Computer').size().div(len(df))
print(prior)

#Step 3 : Likelihood
likelihood = {}
likelihood['Age'] = df.groupby(['Buys_Computer', 'Age']).size().div(len(df)).div(prior)
likelihood['Income'] = df.groupby(['Buys_Computer', 'Income']).size().div(len(df)).div(prior)
likelihood['Student'] = df.groupby(['Buys_Computer', 'Student']).size().div(len(df)).div(prior)
likelihood['CR'] = df.groupby(['Buys_Computer', 'CR']).size().div(len(df)).div(prior)
print (likelihood)

p_yes = (
    likelihood['Age']['yes']['less than 30'] *
    likelihood['Income']['yes']['medium'] *
    likelihood['Student']['yes']['yes'] *
    likelihood['CR']['yes']['fair'] *
    prior['yes']
)

# Probability that the person will NOT buy
p_no = (
    likelihood['Age']['no']['less than 30'] *
    likelihood['Income']['no']['medium'] *
    likelihood['Student']['no']['yes'] *
    likelihood['CR']['no']['fair'] *
    prior['no']
)

print ('Yes : ', p_yes)
print ('No :  ', p_no)

# Step 5: Normalization
print("After Normalization")
v_yes= p_yes/(p_yes+p_no)
print(round(v_yes, 2))

v_no= p_no/(p_yes+p_no)
print(round(v_no, 2))

# Step 6: Final answer

if (v_yes) > (v_no):
    print("For the given query the answer is Yes")
else:
    print("For the given query the answer is No")
```
